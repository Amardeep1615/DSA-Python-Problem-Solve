```python
DSA-
What is DSA?
Data Structures are ways to organize and store the data and it can be used efficiently
Algorithms - Algorithms are step by step instructions to solve a problem

Data Structure + Algorithm = Efficient Problem Solving
(Storage)        (Process)  = Result


Big O Notation 

Time Complexity
Time Complexity measures the how much time an algorithm takes to execute as the input size increases.

Space Complexity
Space Complexity measures the how much memory an alogrithm uses as the input size increases

O(1) - Constant Complexity   - Same time 

O(n) - Linear Time  - Time increases directly with input size

O(n2) - Quardratic Time - Occurs when we use Nested Loops       - In Sorting Algorithms(Bubble Sort,Selection Sort,Insertion Sort)


O(logn) - Logarithmic Time - Time grows slowly as input size increases - Binary Search

O(n logn) - Linearithmic Time - Sorting Algorithms - Merge Sort,Quick Sort,Heap Sort

O(2^n) - Exponential Time  - Very Slow Growth

O(n!) - Factorial Time - Extremely Expensive Complexity - Occurs in problems involving permutations
```

```python
Big-O notation is used to describe the performance or complexity of an algorithm as the input size grows. It helps us understand how efficient an algorithm is in terms of time and space. Common Big-O complexities include O(1) for constant time, O(log n) for logarithmic time, O(n) for linear time, O(n log n) for efficient sorting algorithms, and O(n²) for nested loop operations.
```

```python
Big-O Complexity Order
O(1)
O(log n)
O(n)
O(n log n)
O(n²)
O(2ⁿ)
O(n!)
```

```python
DATA STRUCTURES
│
├── 1️⃣ PRIMITIVE (Basic building blocks)
│   ├── Integer
│   ├── Float
│   ├── Character
│   ├── Boolean
│   └── String
│
├── 2️⃣ LINEAR (Elements in sequential order)
│   │
│   ├── 📦 STATIC (Fixed size)
│   │   └── Array
│   │       ├── 1D Array        → [1, 2, 3, 4]
│   │       ├── 2D Array        → Matrix / Grid
│   │       └── nD Array        → Multi-dimensional
│   │
│   └── 📦 DYNAMIC (Size changes at runtime)
│       │
│       ├── Linked List
│       │   ├── Singly Linked List    → A → B → C → None
│       │   ├── Doubly Linked List    → A ↔ B ↔ C
│       │   └── Circular Linked List → A → B → C → A
│       │
│       ├── Stack
│       │   ├── Array-based Stack
│       │   └── Linked List-based Stack
│       │
│       └── Queue
│           ├── Simple Queue
│           ├── Circular Queue
│           ├── Double-ended Queue (Deque)
│           └── Priority Queue
│
└── 3️⃣ NON-LINEAR (Hierarchical / Networked)
    │
    ├── 🌲 TREES
    │   ├── General Tree
    │   │
    │   ├── Binary Tree
    │   │   ├── Full Binary Tree
    │   │   ├── Complete Binary Tree
    │   │   ├── Perfect Binary Tree
    │   │   └── Skewed Binary Tree
    │   │
    │   ├── Binary Search Tree (BST)
    │   │   └── Self-Balancing BST
    │   │       ├── AVL Tree
    │   │       └── Red-Black Tree
    │   │
    │   ├── Heap
    │   │   ├── Min Heap
    │   │   └── Max Heap
    │   │
    │   ├── Trie (Prefix Tree)
    │   │
    │   ├── Segment Tree
    │   │
    │   ├── Fenwick Tree (Binary Indexed Tree)
    │   │
    │   └── N-ary Tree
    │
    ├── 🌐 GRAPHS
    │   ├── Directed Graph (Digraph)
    │   ├── Undirected Graph
    │   ├── Weighted Graph
    │   ├── Unweighted Graph
    │   ├── Cyclic Graph
    │   ├── Acyclic Graph
    │   └── DAG (Directed Acyclic Graph)
    │
    └── 🗃️ HASHING
        ├── Hash Table
        ├── Hash Map
        └── Hash Set
```

```python
ALGORITHMS
│
├── 1️⃣ SEARCHING
│   ├── Linear Search              → O(n)
│   ├── Binary Search              → O(log n)
│   ├── Ternary Search             → O(log₃ n)
│   ├── Jump Search                → O(√n)
│   ├── Interpolation Search       → O(log log n)
│   └── Exponential Search         → O(log n)
│
├── 2️⃣ SORTING
│   │
│   ├── 🐌 Comparison-Based (Slower)
│   │   ├── Bubble Sort            → O(n²)
│   │   ├── Selection Sort         → O(n²)
│   │   └── Insertion Sort         → O(n²)
│   │
│   └── ⚡ Comparison-Based (Faster)
│       ├── Merge Sort             → O(n log n)
│       ├── Quick Sort             → O(n log n) avg
│       ├── Heap Sort              → O(n log n)
│       ├── Tim Sort               → O(n log n)
│       └── Shell Sort             → O(n log n)
│
│   🚀 Non-Comparison Based (Fastest)
│       ├── Counting Sort          → O(n + k)
│       ├── Radix Sort             → O(nk)
│       └── Bucket Sort            → O(n + k)
│
├── 3️⃣ RECURSION & DIVIDE AND CONQUER
│   ├── Recursion (Base concept)
│   ├── Divide and Conquer
│   │   ├── Merge Sort
│   │   ├── Quick Sort
│   │   └── Binary Search
│   └── Master Theorem
│       (to analyze recursive complexity)
│
├── 4️⃣ GRAPH ALGORITHMS
│   │
│   ├── 🔍 Traversal
│   │   ├── BFS (Breadth First Search)
│   │   └── DFS (Depth First Search)
│   │
│   ├── 🛣️ Shortest Path
│   │   ├── Dijkstra's Algorithm
│   │   ├── Bellman-Ford Algorithm
│   │   └── Floyd-Warshall Algorithm
│   │
│   ├── 🌲 Minimum Spanning Tree
│   │   ├── Kruskal's Algorithm
│   │   └── Prim's Algorithm
│   │
│   └── 🔗 Advanced Graph
│       ├── Topological Sort
│       ├── Tarjan's Algorithm (SCC)
│       ├── Kosaraju's Algorithm
│       └── Union-Find (Disjoint Set)
│
├── 5️⃣ DYNAMIC PROGRAMMING (DP)
│   │
│   ├── Memoization (Top-Down)
│   ├── Tabulation  (Bottom-Up)
│   │
│   └── Classic DP Problems
│       ├── Fibonacci
│       ├── 0/1 Knapsack
│       ├── Longest Common Subsequence (LCS)
│       ├── Longest Increasing Subsequence (LIS)
│       ├── Coin Change
│       ├── Matrix Chain Multiplication
│       └── Edit Distance
│
├── 6️⃣ GREEDY ALGORITHMS
│   ├── Activity Selection
│   ├── Fractional Knapsack
│   ├── Huffman Encoding
│   ├── Job Sequencing
│   └── Greedy Graph Algorithms
│       ├── Dijkstra
│       ├── Prim's
│       └── Kruskal's
│
├── 7️⃣ BACKTRACKING
│   ├── N-Queens Problem
│   ├── Sudoku Solver
│   ├── Rat in a Maze
│   ├── Word Search
│   ├── Subset Sum
│   └── Permutations & Combinations
│
├── 8️⃣ TREE ALGORITHMS
│   ├── Tree Traversals
│   │   ├── Inorder   (Left → Root → Right)
│   │   ├── Preorder  (Root → Left → Right)
│   │   ├── Postorder (Left → Right → Root)
│   │   └── Level Order (BFS on tree)
│   ├── BST Operations
│   ├── LCA (Lowest Common Ancestor)
│   └── Tree DP
│
├── 9️⃣ MATHEMATICAL ALGORITHMS
│   ├── GCD / LCM (Euclidean Algorithm)
│   ├── Sieve of Eratosthenes (Primes)
│   ├── Fast Exponentiation
│   ├── Modular Arithmetic
│   ├── Bit Manipulation
│   └── Number Theory
│
└── 🔟 ADVANCED INTERVIEW PATTERNS
    ├── Two Pointers
    ├── Sliding Window
    ├── Fast & Slow Pointers
    ├── Merge Intervals
    ├── Cyclic Sort
    ├── In-place Reversal (Linked List)
    ├── Tree BFS Pattern
    ├── Tree DFS Pattern
    ├── Two Heaps Pattern
    ├── Subsets Pattern
    ├── Modified Binary Search
    ├── Bitwise XOR
    ├── Top K Elements
    ├── K-way Merge
    ├── 0/1 Knapsack (DP)
    └── Topological Sort Pattern
```

```python
SPEED RANKING (Fastest → Slowest)
─────────────────────────────────
O(1)        → Constant    ⚡⚡⚡⚡⚡  Hash lookup
O(log n)    → Logarithmic ⚡⚡⚡⚡   Binary search
O(√n)       → Square root ⚡⚡⚡    Jump search
O(n)        → Linear      ⚡⚡     Linear search
O(n log n)  → Linearithmic⚡      Merge sort
O(n²)       → Quadratic   🐌      Bubble sort
O(n³)       → Cubic       🐌🐌    Matrix multiply
O(2ⁿ)       → Exponential 🐌🐌🐌  Brute force subsets
O(n!)       → Factorial   🐌🐌🐌🐌 Brute force permutations
```

```python
BEGINNER           INTERMEDIATE         ADVANCED
────────           ────────────         ────────
Arrays          →  Linked Lists      →  Graphs
Strings         →  Trees             →  Dynamic Programming
Hashing         →  BST               →  Advanced Trees
Stack           →  Heap              →  Backtracking
Queue           →  Recursion         →  Greedy
Searching       →  Sorting           →  Interview Patterns
```
